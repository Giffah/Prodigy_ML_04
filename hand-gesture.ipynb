```python
folders_names = []

for i in range(10):
    folder = r"/kaggle/input/leapgestrecog/leapGestRecog/0{}".format(i)
    folders_names.append(folder)

files_names = ['01_palm', '02_l', '03_fist', '04_fist_moved', '05_thumb', '06_index', '07_ok', '08_palm_moved', '09_c', '10_down']
folders_names
        
```

```python
training_data=[]

for folder in folders_names:
        Class_num=folder[-1]
    
        for file in files_names:
            path = os.path.join(folder, file)
            for img in tqdm(os.listdir(path)):
                img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
                
                training_data.append([img_array.flatten(),int(Class_num)])
```

```python
image_data=[]
labels=[]

for feature,label in training_data:
    image_data.append(feature)
    labels.append(label)
```

```python
image_data = np.array(image_data)
labels = np.array(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(image_data, labels, test_size=0.30, random_state=42, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)

```

```python
y_predict = rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_predict,y_test)*100
```
